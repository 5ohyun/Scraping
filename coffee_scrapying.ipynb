```python
import requests
from bs4 import BeautifulSoup
import time
from selenium import webdriver as wd
import pandas as pd
import numpy as np
```

```python
coffee_list = []


response = requests.get('https://www.baristapaulbassett.co.kr/menu/List.pb?cid1=D')

# A : Coffee / B:Becerage / C: Ice-Cream / D: Food / E : Product

soup = BeautifulSoup(response.text, 'html.parser')

link = soup.select('.listStyleB > li > a')
img = soup.select('.listStyleB > li > a > div > img')
kor_name = soup.select('.listStyleB > li > a > div.txtArea')
eng_name = soup.select('.listStyleB > li > a > div.txtArea > span')
```

```python
for one in range(len(link)) :
    link_one=link[one]["onclick"][8:16]
    img_link_one='https://www.baristapaulbassett.co.kr'+img[one]['src']
    kor_name_one = kor_name[one]
    kor_name_one = str(kor_name_one).split('<')[1].split('>')[1]
    eng_name_one = eng_name[one].get_text()
    coffee_list.append([link_one, img_link_one, kor_name_one,eng_name_one])
```

```python
coffee_list=pd.DataFrame(coffee_list, columns = ['link','img','kor','eng'])
coffee_list
```

```python
coffee_list.to_csv('food_list.csv',index=False)
```

```python
# <COFFEE_INFORMATION>

coffee_info = []
for i in coffee_list['link'] :
    response = requests.get('https://www.baristapaulbassett.co.kr/menu/View.pb?dpid='+i)
    soup = BeautifulSoup(response.text, 'html.parser')

    summary = soup.select_one('.menuInfo > .menuTit > dl > dd').get_text()

    # 0.icehot 1. milk 2. allergy 3. size 

    if len(soup.select('.menuInfo > .info > li')) == 1 : # 알러지
      icehot = np.nan
      milk = np.nan
      allergy = str(soup.select('.menuInfo > .info > li')[0].get_text()).replace(str(soup.select('.menuInfo > .info > li > span')[0].get_text()),'').strip()
      size = np.nan

    elif len(soup.select('.menuInfo > .info > li')) == 2 : # ice, 사이즈
      icehot = str(soup.select('.menuInfo > .info > li')[0].get_text()).replace(str(soup.select('.menuInfo > .info > li > span')[0].get_text()),'').strip()
      milk = np.nan
      allergy = np.nan
      size = re.sub(r"\s", "",str(soup.select('.menuInfo > .info > li')[1].get_text()).replace(str(soup.select('.menuInfo > .info > li > span')[1].get_text()),''))

    elif len(soup.select('.menuInfo > .info > li')) == 3 : ## ice, 사이즈, 알러지
      icehot = str(soup.select('.menuInfo > .info > li')[0].get_text()).replace(str(soup.select('.menuInfo > .info > li > span')[0].get_text()),'').strip()
      milk = np.nan
      allergy = str(soup.select('.menuInfo > .info > li')[1].get_text()).replace(str(soup.select('.menuInfo > .info > li > span')[1].get_text()),'').strip()
      size = re.sub(r"\s", "",str(soup.select('.menuInfo > .info > li')[2].get_text()).replace(str(soup.select('.menuInfo > .info > li > span')[2].get_text()),''))

    elif len(soup.select('.menuInfo > .info > li')) == 0 :
      icehot = np.nan
      milk = np.nan
      allergy = np.nan
      size = np.nan

    else :
      icehot = str(soup.select('.menuInfo > .info > li')[0].get_text()).replace(str(soup.select('.menuInfo > .info > li > span')[0].get_text()),'').strip()
      milk = re.sub(r"\s", "",str(soup.select('.menuInfo > .info > li')[1].get_text()).replace(str(soup.select('.menuInfo > .info > li > span')[1].get_text()),'').strip())
      allergy = str(soup.select('.menuInfo > .info > li')[2].get_text()).replace(str(soup.select('.menuInfo > .info > li > span')[2].get_text()),'').strip()
      size = re.sub(r"\s", "",str(soup.select('.menuInfo > .info > li')[3].get_text()).replace(str(soup.select('.menuInfo > .info > li > span')[3].get_text()),''))

    coffee_info.append([i,summary, icehot, milk,allergy,size])
```

```python
coffee_info=pd.DataFrame(coffee_info, columns = ['link','summary','icehot','milk','allergy','size'])
coffee_info
```

```python
## <COFFEE_NUTRITION>

coffee_info = []
for i in coffee_list['link'] :
    response = requests.get('https://www.baristapaulbassett.co.kr/menu/View.pb?dpid='+i)
    soup = BeautifulSoup(response.text, 'html.parser')

    gml = soup.select_one('.menuInfo > .nutritional > .txtArea > div > .sizeMl').get_text()
    # 0 : kcal 1. sugar 2.salt 3. protein 4.fat 5.caffe
    kcal = soup.select('.menuInfo > .nutritional > .txtArea > div')[0].select('ul > li > .num')[0].get_text()
    sugar = soup.select('.menuInfo > .nutritional > .txtArea > div')[0].select('ul > li > .num')[1].get_text()
    salt = soup.select('.menuInfo > .nutritional > .txtArea > div')[0].select('ul > li > .num')[2].get_text()
    protein = soup.select('.menuInfo > .nutritional > .txtArea > div')[0].select('ul > li > .num')[3].get_text()
    fat = soup.select('.menuInfo > .nutritional > .txtArea > div')[0].select('ul > li > .num')[4].get_text()
    caffe = soup.select('.menuInfo > .nutritional > .txtArea > div')[0].select('ul > li > .num')[5].get_text()

    coffee_info.append([i,gml, kcal, sugar,salt,protein,fat,caffe])
```

```python
coffee_info=pd.DataFrame(coffee_info, columns = ['link','gml','kcal','sugar','salt','protein','fat','caffe'])
coffee_info
```

```python
coffee_info.to_csv('detail_data/food_nutri.csv',index=False)
```
